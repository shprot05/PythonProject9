**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
a = np.random.randint(1,10,size = 10)
print(a)
a[(a>=3) & (a<=8)] *= -1
print(a)
```

```python
b = np.random.randint(1,10,size = 10)
print(b)
max = np.argmax(b)
b[max] = 0
print(b)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
matrix  = np.random.randint(0,5,size = (2,2))
print(f"""Матрица:
{matrix}""")
array = matrix.flatten()
print(f"Массив из матрицы: {array}")
ready_matrix = np.array(np.meshgrid(array,array)).T.reshape(-1,2)
print(f"""Массив из матрицы:
{ready_matrix}""")
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A = np.random.randint(1,6,size = (8,3))
B = np.random.randint(1,6,size = (2,2))
mask = np.all([
    np.isin(A, B[i]).sum(axis=1) == B.shape[1]
    for i in range(B.shape[0])], axis=0)

result = A[mask]
print(result)
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
new_matrix = np.random.randint(1,4, size = (10,3))
print(new_matrix)
result = []
for row in new_matrix:
    if row[0] != row[1] or row[1] != row[2]:
        result.append(row)
ready_matrix = np.array(result)
print("")
print(ready_matrix)
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

```python
matrix = np.random.randint(1, 3, size=(10, 3))
unique_rows = []

for row in matrix:
    if list(row) not in unique_rows:
        unique_rows.append(list(row))

result = np.array(unique_rows)
print(result)
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
diagonal = np.diag(X)
diagonal = diagonal[diagonal != 0]
result = np.prod(diagonal)
print(f'c num py: {result}')



# без num.py
Y = [[1, 0, 1],
     [2, 0, 2],
     [3, 0, 3],
     [4, 4, 4]]

diagonal = []
for i in range(min(len(Y), len(Y[0]))):
    if Y[i][i] != 0:
        diagonal.append(Y[i][i])

result = 1
for num in diagonal:
    result *= num

print(f'без нам пай: {result}')
```
