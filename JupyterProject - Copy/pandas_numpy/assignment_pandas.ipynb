```python
# загружаем датасет
data = pd.read_csv("./data/adult.data.csv")
data.head()
```

**1. Посчитайте, сколько мужчин и женщин (признак *sex*) представлено в этом датасете**

```python
print(data['sex'].value_counts())
```

**2. Каков средний возраст мужчин (признак *age*) по всему датасету?**

```python
print(data[data['sex'] == 'Male']['age'].mean())
```

**3. Какова доля граждан Соединенных Штатов (признак *native-country*)?**

```python
us_share = (data['native-country'] == 'United-States').mean()
print(f'{us_share:.2%}')
```

**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
low_income = data[data['salary'] == '<=50K']['age']
print(f"Средний возраст (<=50K): {low_income.mean():.1f}")
print(f"Ст. отклонение (<=50K): {low_income.std():.1f}")


high_income = data[data['salary'] == '>50K']['age']
print(f"Средний возраст (>50K): {high_income.mean():.1f}")
print(f"Ст. отклонение (>50K): {high_income.std():.1f}")
```

**6. Правда ли, что люди, которые получают больше 50k, имеют минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
result = data[data['salary'] == '>50K']['education'].isin([
    'Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']).all()

print(result)
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Asian-Pac-Islander*.**

```python
print(data.groupby(['race','sex'])['age'].describe())
max_age = data[
(data['race'] == 'Asian-Pac-Islander') &
    (data['sex'] == 'Male')]['age'].max()

print(f"Максимальный возраст мужчин Asian-Pac-Islander: {max_age}")
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
married_men = data[
    (data['sex'] == 'Male') &
    (data['marital-status'].isin(['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']))
]

single_men = data[
    (data['sex'] == 'Male') &
    (~data['marital-status'].isin(['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']))
]

married_rich = (married_men['salary'] == '>50K').mean()
single_rich = (single_men['salary'] == '>50K').mean()

print(f"Женатые мужчины (>50K): {married_rich:.2%}")
print(f"Холостые мужчины (>50K): {single_rich:.2%}")
print("Больше зарабатывают женатые мужчины:", married_rich > single_rich)
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours = data['hours-per-week'].max()
print(max_hours)

workers = data[data['hours-per-week'] == max_hours]
rich_share = (workers['salary'] == '>50K').mean()
print(f"Процент зарабатывающих >50K среди них: {rich_share:.2%}")

print(f'Количество человек с максимальным количеством часов работы: {workers.shape[0]}')
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
country_working_time = data.groupby(['native-country', 'salary'])['hours-per-week'].mean()
print(country_working_time)
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
data['AgeGroup'] = data['age'].apply(
    lambda x: 'young' if 16 <= x <= 35 else (
        'adult' if 35 < x <= 70 else (
            'retiree' if 70 < x <= 100 else 'unknown')))

```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
high_income_counts = data[data['salary'] == '>50K'].groupby('AgeGroup').size()
print(high_income_counts)
print(high_income_counts.idxmax())
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
data = pd.DataFrame({
    'occupation': ['Programmer', 'Programmer', 'Teacher', 'Teacher', 'Artist', 'Artist'],
    'age': [25, 35, 45, 50, 22, 23],
    'hours-per-week': [40, 40, 40, 40, 4, 10]
})

def check_conditions(group):
    age_condition = group['age'].mean() <= 40
    hours_condition = (group['hours-per-week'] > 5).all()
    return age_condition and hours_condition

group_checks = data.groupby('occupation')[['age', 'hours-per-week']].apply(check_conditions)

print("Результат проверки каждой группы:")
print(group_checks)

# Фильтруем список профессий
valid_occupations = group_checks[group_checks].index.tolist()
print(f"\nТипы занятости: {valid_occupations}")
```
